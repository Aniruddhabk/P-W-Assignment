## Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
### Ans: Glyphs in Bokeh are graphical primitives used to represent visualizations of data. They can be used to create a wide variety of plots such as lines, circles, rectangles, and more. Bokeh provides a wide range of built-in glyphs that can be customized as per your requirement.To add glyphs to a Bokeh plot, you can use the glyph methods provided by the figure object.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure ,output_file , show
```

```python


# create a figure object
plot = figure(title='Glyphs Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# add circle glyph to the plot
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
plot.circle(x, y, size=10, color='red', alpha=0.5)

# set the output file name
output_file('bokeh_glyphs.html')

# show the plot
show(plot)
```

```python
# create a figure object
plot = figure(title='Glyphs Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# add line and circle glyphs to the plot
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
plot.line(x, y, line_width=2, color='blue', alpha=0.8)
plot.circle(x, y, size=10, color='red', alpha=0.5)

# set the output file name
output_file('bokeh_multiple_glyphs.html')

# show the plot
show(plot)
```

## Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?
### Ans:A Bokeh server is a Python process that allows you to create and serve Bokeh plots and applications over the web. With a Bokeh server, you can create interactive plots and applications that can be updated in real-time, as users interact with them.To create a Bokeh server application, you first define the layout and behavior of the application using Python code, just likyou would for a standalone Bokeh plot. You can define callbacks that respond to user input or changes to the data in the plotand update the plot or other elements of the application in response.Once you have defined the application, you can start the Bokeh server and publish the application to the web, where users caaccess it through a web browser. The Bokeh server provides a number of tools for interacting with and debugging youapplications, including a web-based console and a real-time data streaming interface.To use the Bokeh server, you will need to install the bokeh package and start the server using the bokeh serve command in terminal window. You can then access the application by navigating to the URL 
###### http://localhost:5006/<appname>,where appname ithe name of the application you defined.

```python
from bokeh.io import curdoc
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.server.server import Server
from bokeh.application import Application
from bokeh.application.handlers.function import FunctionHandler
from random import random

# define a function that creates and returns a Bokeh plot
def make_plot(doc):
    # create a new plot with a random line
    source = ColumnDataSource(data=dict(x=[1, 2, 3, 4, 5], y=[random() for i in range(5)]))
    plot = figure(plot_height=300, plot_width=600, title="Streaming Line Plot")
    line = plot.line(x='x', y='y', source=source)

    # define a callback that updates the line data every second
    def update():
        new_data = dict(x=[i+1 for i in source.data['x']], y=[random() for i in range(5)])
        source.data = new_data

    # add the callback to the document
    doc.add_periodic_callback(update, 1000)

    # add the plot to the document
    doc.title = "Streaming Line Plot"
    doc.add_root(plot)

# create a new Bokeh application with the make_plot function as the handler
app = Application(FunctionHandler(make_plot))

# create a new Bokeh server with the application and start it
server = Server({'/': app})
server.start()

# publish the plot to a web page
server.show('/')

# run the Bokeh server
server.io_loop.start()
```
